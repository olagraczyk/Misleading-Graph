This is one of the many examples used by media worldwide to manipulate viewers into perceiving a significant difference between events. This particular graph was shown during TVP1’s evening news program. At first glance, the viewer might think that TVP2 and TVP1 are far ahead of other popular TV stations. However, upon closer examination of the actual numbers, it becomes clear that these differences are minimal—not even a full percentage point.  

This manipulation was achieved by setting the y-axis to start from a specific number, in this case, 7.0% up to 8.0%, instead of starting from 0.0%.  

Unfortunately, this is one of the most common practices in media. While it is technically not lying, it exploits human perception to exaggerate differences. This kind of manipulation could be easily detected and corrected using AI tools, which could not only identify the issue but also adjust the y-axis automatically, even without a database.

###5. Create an improved chart (preferably using ggplot2, Pandas, or Seaborn, but any tool is acceptable) – include both code and result



````python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

#Estimated values extracted by AI tool from the graph
data = pd.DataFrame({
    "TV Station": ["TVP2", "TVP1", "TVN", "Polsat"],
    "Viewership (%)": [7.85, 7.75, 7.30, 7.15]
})

# I am using simple style and colors
sns.set_style("whitegrid")
plt.figure(figsize=(8, 5))

# Create a bar plot
ax = sns.barplot(
    x="TV Station",
    y="Viewership (%)",
    hue="TV Station",
    data=data,
    palette=["lightgreen"]*len(data),
)

# Add values on top of bars
for index, value in enumerate(data["Viewership (%)"]):
    plt.text(index, value + 0.05, f"{value:.2f}%", ha='center', fontsize=12)

# Adjust axis
ax.set(xlabel=None)

plt.ylim(0, 10) # Adjust y-axis to start from 0

plt.title("UDZIAŁY 4 GŁÓWNYCH STACJI W MARCU 2023", fontsize=14)
plt.ylabel("Oglądalność (%)", fontsize=12)

# Display the plot
plt.show()
````
